```python
def std_date_of_birth(date):
    return date.split(" ")[0]

def std_height(height):
    return height.split("c")[0]

df_midfielder['Date of Birth'] = df_midfielder['Date of Birth'].apply(std_date_of_birth)
df_midfielder['Height'] = df_midfielder['Height'].apply(std_height)
df_midfielder
```

```python
# Identify duplicates based on 'Name'
duplicates = df_midfielder.sort_values(by='Appearances').duplicated(subset='Name',keep='first')

# Remove all duplicate rows from the original DataFrame
df_midfielder = df_midfielder[~duplicates]
```

```python
df_midfielder = df_midfielder.dropna()
```
