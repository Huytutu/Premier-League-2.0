```python
import pandas as pd
```

```python
df_goalkeeper = pd.read_csv('../../data/Stats_csv/Forward_stats.csv')
```

```python
def std_date_of_birth(date):
    return date.split(" ")[0]

def std_height(height):
    return height.split("c")[0]

df_goalkeeper['Date of Birth'] = df_goalkeeper['Date of Birth'].apply(std_date_of_birth)
df_goalkeeper['Height'] = df_goalkeeper['Height'].apply(std_height)
df_goalkeeper
```

```python
# Identify duplicates based on 'Name'
duplicates = df_goalkeeper.sort_values(by='Appearances').duplicated(subset='Name',keep='first')

# Remove all duplicate rows from the original DataFrame
df_goalkeeper = df_goalkeeper[~duplicates]
```

```python
df_goalkeeper = df_goalkeeper.dropna()
```

```python
df_goalkeeper.to_csv('../../data/Stats_csv/Processed_Goalkeeper_data.csv',index=False) 
```
